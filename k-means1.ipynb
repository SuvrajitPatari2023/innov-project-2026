## Using the chipotle_stores.csv dataset
---

```python
df: pd = pd.read_csv("chipotle_stores.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['latitude'], df['longitude'], marker="*")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1, 15):
    KM = KMeans(n_clusters=i, random_state=42)
    KM.fit(df[['latitude', 'longitude']])
    wcss.append(KM.inertia_)
```

```python
plt.plot(range(1, 15), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
X = df[['latitude', 'longitude']].iloc[:, :].values
km = KMeans(n_clusters=3)
y_means = km.fit_predict(X)
```

```python
for i in range(3):
    plt.scatter(X[y_means == i, 0], X[y_means == i, 1], marker='*')

# plt.legend(['cluster 0', 'cluster 1', 'cluster 2'])
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X, y_means)
```

## Using the zomato_restaurants_in_India.csv dataset
---

```python
df: pd = pd.read_csv("zomato_restaurants_in_India.csv")
df.head()
```

```python
df[['latitude', 'longitude']].max()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['latitude'], df['longitude'], marker="*")
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1, 15):
    KM = KMeans(n_clusters=i, random_state=42)
    KM.fit(df[['latitude', 'longitude']])
    wcss.append(KM.inertia_)

plt.plot(range(1, 15), wcss)
plt.show()
```

```python
X = df[['latitude', 'longitude']].iloc[:, :].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
for i in range(4):
    plt.scatter(X[y_means == i, 0], X[y_means == i, 1], marker='*')

plt.legend(['cluster 0', 'cluster 1', 'cluster 2', 'cluster 3'])
plt.xlabel('LATITUDE')
plt.ylabel('LONGITUDE')
plt.show()
```
