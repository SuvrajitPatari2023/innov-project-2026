# K-means clustering
---

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student_clustering.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'], df['iq'])
plt.xlabel('cgpa ->')
plt.ylabel('iq ->')
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1, 15):
    KM = KMeans(n_clusters=i)
    KM.fit(df)
    wcss.append(KM.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 15), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
X = df.iloc[:, :].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
X[y_means == 3, 0]
```

```python
for i in range(4):
    plt.scatter(X[y_means == i, 0], X[y_means == i, 1], marker='*')

plt.legend(['cluster 0', 'cluster 1', 'cluster 2', 'cluster 3'])
# plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1])
# plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1])
# plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1])
```
