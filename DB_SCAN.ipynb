```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
X_small = np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[56,64],[25,80]])

plt.figure()
plt.scatter(X_small[:,0], X_small[:,1])
plt.title("Example Dataset")
plt.show()
```

```python
model_small = DBSCAN(eps=3, min_samples=2)
labels_small = model_small.fit_predict(X_small)

plt.figure()
plt.scatter(X_small[:,0], X_small[:,1], c=labels_small)
plt.title("DBSCAN on Example Dataset")
plt.show()

labels_small
```

```python
df = pd.read_csv("chipotle_stores.csv")
df.head()
```

```python
X = df.select_dtypes(include=['int64', 'float64']).values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = DBSCAN(eps=0.3, min_samples=70)
labels = model.fit_predict(X_scaled)

labels[:10]
```

```python
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")
plt.show()
```

```python
silhouette_score(X_scaled, labels)
```
